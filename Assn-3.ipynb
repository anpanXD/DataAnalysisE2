### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

plt.figure()
plt.plot(a, color = 'black', marker = 'o')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
b = np.load( 'data-fig2.npy' )
b1 = b.T[0]
b2 = b.T[1]
b3 = b.T[2]

plt.figure()
plt.plot(b1, color = 'red', label = 'Red')
plt.plot(b2, color = 'green', label = 'Green')
plt.plot(b3, color = 'blue', label = 'Blue')
plt.legend()
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c = np.load('data-fig3.npy')

c1 = c.T[0]
c2 = c.T[1]

plt.figure()
plt.scatter(c1, c2, color = 'blue', alpha = .27, s = 200)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d = np.load('data-fig4.npy')
dx = np.array([1, 2, 3, 4, 5])
xlabels = ['A', 'B', 'C', 'D', 'E']

plt.figure()
plt.bar(dx, d, color = 'grey', alpha = .6)
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks( ticks = dx, labels = xlabels)
plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e = np.load('data-fig5.npy')

plt.figure()
plt.pcolormesh(e, vmin = -1, vmax = 1, cmap = 'RdBu')
plt.colorbar()
plt.show()
```
